```python
# load the list of detections 

# detectionList maps a filename to a list of numbers:
# the first 4 integers are the left, top, right, and bottom (respectively) coordinates of the detected bounding box
# the next 5 numbers are in the format M:N where M is the tag ID and N is the confidence of the classification (top 5)
# the classifications are sorted in decreasing order of confidence

detections = pickle.load(open(DATA_BASE_PATH + 'detectionsList_classified.pckl', 'rb'), encoding="latin1")
```

```python
# make a map of tag ids to human-readable labels
tag_list_file = '29kfiles/map_labels_tags.txt'
with open(DATA_BASE_PATH + tag_list_file) as file:  
    tag_list = file.read() 
tag_list = tag_list.split('\n')
tag_map = dict((int(elem.split(' ',1)[0]),elem.split(' ',1)[1]) for elem in tag_list if elem) # map tag ID to name
```

```python
tag_map
```

```python
sample = None
for key, val in detections.items(): 
    print(key)
    print(val)
    sample = val
    break
```

```python
def decode_detection(elt): 
    # given one element from the vals of detections, convert it into a list of classification ids 
    # and a list of confidences 
    bboxnums = elt.split()
    tags = []
    conf = [] 
    
    for tag in bboxnums[5:]: 
        #print(tag.split(":"))
        #print([float(num) for num in tag.split(":")])
        t = [float(num) for num in tag.split(":")]
        tag_id = int(t[0])
        tag_conf = t[1]
        #print("tag_id", tag_id)
        #print("tag_conf", tag_conf)
        tags.append(tag_id)
        conf.append(tag_conf)
    return tags, conf

#     # convert coordinates of bounding box from normalized coordinates to rectangle coordinates
#     left = int(w*float(bboxnums[0])/1000.0)
#     top = int(h*float(bboxnums[1])/1000.0)
#     right = int(w*float(bboxnums[2])/1000.0)
#     bottom = int(h*float(bboxnums[3])/1000.0)
```

```python
print(sample[0])
decode_detection(sample[0])
```

```python
def get_icon_topics(detection): 
    all_tags = set()
    for icon in detection: 
        tags, conf = decode_detection(icon)
        assert len(tags) == len(conf)
        for i in range(len(conf)): 
            if conf[i] >= .1: 
                all_tags.add(tags[i])
    return all_tags
    
```

```python
get_icon_topics(sample)
```

```python
# calculate valid icons for all tags 
icons = {} 
for key, val in detections.items(): 
    icons[key] = list(get_icon_topics(val))
```

```python
f_icons = {}
for key, val in icons.items(): 
    if len(val) > 0: 
        f_icons[key] = val
len(f_icons)
```

```python
icon_texts = []
for key, val in f_icons.items(): 
    icon_texts.append([str(elt) for elt in val])
```

```python
icon_text_d = Dictionary(icon_texts)
icon_text_d.filter_extremes(no_below=10, no_above=.5)
icon_id2token = dict([(id_, token) for token, id_ in icon_text_d.token2id.items()])
```

```python
common_corpus = [icon_text_d.doc2bow(text) for text in icon_texts]
```

```python
lda = LdaModel(common_corpus, num_topics=N_TOPICS_ICONS)
```

```python
for t_id in range(N_TOPICS_ICONS): 
    terms = lda.get_topic_terms(t_id, 20)
    print("Topic: %d" % t_id)
    cats = []
    for t in terms: 
        icon_id = icon_id2token[t[0]]
        icon_cat_name = tag_map[int(icon_id)]
        cats.append(icon_cat_name)
    print(cats)
```

# Test a function for cropping bboxes

```python
detectionsList = pickle.load(open(DATA_BASE_PATH + "detectionsList_classified.pckl", "rb"))

def crop_bbox(info, bbox_idx): 
    print("info", info)
    bbox = detectionsList[info][bbox_idx]
    base_path = "test/final/"
    im = Image.open(base_path + info)
    w,h = im.size
    bboxnums = bbox.split()
    # convert coordinates of bounding box from normalized coordinates to rectangle coordinates
    left = int(w*float(bboxnums[0])/1000.0)
    top = int(h*float(bboxnums[1])/1000.0)
    right = int(w*float(bboxnums[2])/1000.0)
    bottom = int(h*float(bboxnums[3])/1000.0)
    im = im.crop((left, top, right, bottom))
    return im
```

```python
def fix_png(img): 
    if img.mode == 'RGBA':       
        print("Fixing")
        img.load()
        bg = Image.new("RGB", img.size, (255, 255, 255))
        bg.paste(img, mask=img.split()[3]) 
        img = bg
    return img
```

# Lda images based on clusters

```python
# load cluster map 
cluster_map = json.load(open('cluster_map.json', 'r'))
```

```python
N_TOPICS_ICON_FEATS = 20
```

```python
# each infographic should be a list of cluster ids 
doc_map = {}
for cluster_id, bbox_info in cluster_map.items(): 
    for b in bbox_info: 
        info, b_idx = b 
        doc_map[info] = doc_map.get(info, []) + [cluster_id]

#flatten 
doc_map_flat = [(key, val) for key, val in doc_map.items()]
docs = [elt[1] for elt in doc_map_flat]
```

```python
icon_feat_d = Dictionary(docs)
icon_feat_d.filter_extremes(no_below=10, no_above=.5)
icon_feat_id2token = dict([(id_, token) for token, id_ in icon_feat_d.token2id.items()])
```

```python
common_corpus = [icon_feat_d.doc2bow(text) for text in docs]
```

```python
lda = LdaModel(common_corpus, num_topics=N_TOPICS_ICON_FEATS)
```

```python
human_bboxes = pickle.load(open(VISUALLY_LINK + HUMAN_BBOXES, 'rb'))
all_bboxes = pickle.load(open(VISUALLY_LINK + BBOXES, 'rb'))
```

```python


def crop_bbox_all(info, bbox_idx): 
    bbox = all_bboxes[info][bbox_idx]
    im = Image.open(INFO_PATH + info)
    w,h = im.size
    bboxnums = bbox.split()
    # convert coordinates of bounding box from normalized coordinates to rectangle coordinates
    left = int(w*float(bboxnums[0])/1000.0)
    top = int(h*float(bboxnums[1])/1000.0)
    right = int(w*float(bboxnums[2])/1000.0)
    bottom = int(h*float(bboxnums[3])/1000.0)
    im = im.crop((left, top, right, bottom))
    return im

def crop_bbox_humans(info, bbox_idx): 
    box = human_bboxes[info][bbox_idx]
    im = Image.open(INFO_PATH + info)
    w,h = im.size

    bottom = int(box['bottom']*h)
    top = int(box['top']*h)
    left = int(box['left']*w)
    right = int(box['right']*w)
    # hack! 
    if left == right: 
        right = right + 1
    if top == bottom: 
        top = top + 1
    #print("bottom-top", bottom-top)
    #print("left-right", left-right)

    im = im.crop((left, top, right, bottom))
    print("crop size", im.size)
    return im

def print_cluster(n_cluster, n_samples, is_human=False, should_display=True): 
    crop_fn = crop_bbox_humans if is_human else crop_bbox_all
    samples = cluster_map[n_cluster]
    random.shuffle(samples)
    images = []
    for i in range(min(n_samples, len(samples))): 
        try: 
            info, bbox = samples[i]
            im = crop_fn(info, bbox)
            images.append(im)
            #stem = "png" if im.mode == "RGBA" else "jpg"
            #if should_display: 
                #display(im)
            #im_name = "cluster_%d_idx_%d.%s" % (n_cluster, i, stem)
            #im.save(EXAMPLE_CLUSTER_FOLDER + im_name)
        except Exception as e: 
            print(e)
    if should_display: 
        showImagesHorizontally(images)
```

```python
def showImagesHorizontally(images):
    fig = figure()
    for i in range(len(images)):
        a=fig.add_subplot(1,len(images),i+1)
        #image = imread(list_of_files[i])
        imshow(images[i])
        axis('off')
```

```python
def print_topic(topic_num, top_clusters): 
    print("TOPIC NUMBER %d" % topic_num)
    for i, c in enumerate(top_clusters): 
        print("Cluster #%d: %s" % (i, c))
        print_cluster(c, 3, is_human=False, should_display=True)
```
