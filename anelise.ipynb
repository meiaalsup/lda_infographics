```python
import pickle
```

```python
DATA_BASE_PATH = "../visuallydata/"
SAMPLE_INFOGRAPHIC = "road-traffic-accidents-infographic_50290aaae47b1.png"
```

```python
# with open(DATA_BASE_PATH +'raw_ocr_output.pickle','rb') as infile:
#     raw_ocr_output = pickle.load(infile)
```

```python
with open(DATA_BASE_PATH +'google_text_extraction_output.pckl','rb') as infile:
    google_text_extraction = pickle.load(infile)
```

```python
texts = [] 
for key, val in google_text_extraction.items(): 
    texts.append([s.lower() for s in val])
    if (len(texts) > 100):
        break
texts[0]
```

```python
from gensim.test.utils import common_texts
from gensim.corpora.dictionary import Dictionary
from gensim.models.ldamodel import LdaModel
```

```python
text_d = Dictionary(texts)
id2token = dict([(id_, token) for token, id_ in text_d.token2id.items()])
```

```python
id2token
```

```python
common_corpus = [text_d.doc2bow(text) for text in texts]
```

```python
N_TOPICS = 10
```

```python

lda = LdaModel(common_corpus, num_topics=N_TOPICS)
```

```python
print([token for id_, token in text_d.id2token.items()])
```

```python
text_d.token2id
```

```python
for t_id in range(N_TOPICS): 
    terms = lda.get_topic_terms(t_id, 20)
    print("Topic: %d" % t_id)
    for t in terms: 
        print("\t%s" % (id2token[t[0]]))
```
